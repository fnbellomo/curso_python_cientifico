```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def presion_gas_ideal(moles, temperatura, volumen):
    r = 8.314
    return (moles * r * temperatura) / volumen
```

```python
n_step = 10
mole = np.random.random_integers(1, 25, size = 50)
T = np.random.random_integers(250, 300, size = 50)
vol = np.linspace(1, 10, num=n_step)

x_out = np.zeros(2)
for i in range(n_step):
    x_out[0] = vol[i]
    x_out[1] = presion_gas_ideal(mole[0], T[0], vol[i])
    print x_out
    np.savetxt('data_input/datos_3*/pru.txt', x_out, fmt='%.4f')
```
